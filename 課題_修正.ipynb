都道府県をそれぞれの平均気温（℃）と年間降水量（mm）の値を元に散布図で可視化

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/jmorigit/intro_to_python/main/weather_data.csv', encoding='utf-8')
plt.figure(figsize=(10,10))
plt.scatter(df['temp'], df['rain'], s = 10)
plt.xlabel('avg. temp.')
plt.ylabel('rainfall')
plt.grid(True)
for i, j, label in zip(df['temp'], df['rain'], df['pref']):
        plt.text(i, j, label, fontsize = 'small')
plt.show()
```
